```python
import heapq
import math

def dijkstra_shortestPath (graph, first_vertex, last_vertex):
    
    distances = {vertex: float('infinity') for vertex in graph}
    distances[first_vertex] = 0
    
    pq = [(0, first_vertex)]
    
    prev_verteces = {vertex: None for vertex in graph}
    
    while pq:
        current_distance, current_vertex = heapq.heappop(pq)
        
        if current_vertex == last_vertex:
            
            path = []
            
            while prev_verteces[current_vertex] is not None:
                path.insert(0, current_vertex)
                current_vertex = prev_verteces[current_vertex]
            path.insert(0,first_vertex)
            return path, distances[last_vertex]
        
        for neighbor, weight in graph[current_vertex].items():
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                prev_verteces[neighbor] = current_vertex
                heapq.heappush(pq, (distance, neighbor))
                
first_vertex = 'A'
last_vertex = 'S'
shortest_path, shortest_distance = dijkstra_shortestPath(graph, first_vertex, last_vertex)
dijkstra_time = %timeit -o dijkstra_shortestPath
```
